```python
new_df = df.dropna()

# The original DataFrame remains unchanged, we can change the original DataFrame by using inplace=True.
```

```python
df.dropna(inplace=True)
```

```python
df.fillna(130, inplace=True) # It replaces all empty cells in the whole DataFrame.
```

```python
df['cement'].fillna(130, inplace=True) # It only replace empty values for one column, specify the column name.
```

```python
# We can replace with mean median or mode.

x = df['cement'].mean() # or .mode()[0] or .median()
print(x)
df['cement'].fillna(x, inplace=True)
```

```python
# Remove rows with a null value in the cement column.

df.dropna(subset=['cement'], inplace=True)
```

```python
# Replacing values.

df.loc[7, 'cement'] = 140
```

```python
print(df['cement'][7])
```

```python
for x in df.index:
    if df.loc[x, 'cement'] > 400:
        print(df['cement'][x])
        df.loc[x, 'cement'] = 400
```

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
```
