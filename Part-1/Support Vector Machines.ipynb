SVM can perform linear or non-linear classification, regression, and even outlier detection. It draws a line or a hyperplane that separates these points into different classes. It finds the best possible line/hyperplane by maximizing the margin, which is the distance between the line and the nearest data points from each class. 

#### Hard Margin Classification

It is a concept in SVM where the algorithm strictly enforces the margin between classes to be as wide as possible while ensuring that no data points lie within the margin or on wrong side of the decision boundary(hyperplane). 
While hard margin classification can yield a very accurate decision boundary when the data is perfectly separable, it can be sensitive to outliers or noisy data points. 


#### Soft Margin Classification
Unlike hard margin classification, it allows for some misclassification of training examples in order to find a more generalized decision boundary. The objective is to find a balance between maximizing the margin and minimizing the classification errors. This is achieved by introducing a penalty parameter C that controls the trade-off between margin width and the number of misclassified examples. 

A smaller C value leads to a wider street but more margin violations. 
A higher C value leads to a smaller street but less margin violations.

If the SVM model is overfitting, we can try regularizing it by reducing C.

#### Adding Similarity Features

Similarity Function measures how much each instance resembles a particular landmark.

We can use kernel function to transform n-dimensional input to m-dimensional input, where m >> n. The main idea to use kernel is: A linear classifier or regression curve in higher dimensions become a non-linear classifier or regression curve in lower dimensions.


Radial Basis Function Kernel, also called the Gaussian kernel operates by measuring the similarity between data points based on their Euclidean distance in the input space. 
 ϕγ(x,ℓ) = exp(−γ∥x−ℓ∥^2)

We can create a landmark at the location of each and every instance in the dataset. This creates many dimensions and thus increases the chances that the transformed training set will be linearly separable.

```python
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
    

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
```

```python
rbf_kernel_svm_clf = Pipeline((
                        ("scaler", StandardScaler()),
                        ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
))
rbf_kernel_svm_clf.fit(X, y)

plot_predictions(rbf_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

#### Computational Complexity

LinearSVC class is based on the liblinear library, its training time complexity is roughly O(m * n). The algorithm takes longer if we required a very high precision, which can be controlled by the tolerance hyperparameter called tol in sklearn. 

SVM class is based on libsvm library, which implements an algorithm that supports the kernel trick. Its time complexity is usually between O(m^2 * n) and O(m^3 * n). So, it becomes very slow when the number of instances gets large. The algorithm is perfect for complex but small or medium training sets.

### SVM Regression

It tries to fit as many instances as possible on the street while limiting margin voilations. The width of the street is controlled by a hyperparameter epsilion. Large epsilion means large width and vice versa.

```python
from sklearn.svm import LinearSVR

np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3*X + np.random.randn(m, 1)).ravel() # follows normal(Guassian) Distribution
svm_reg = Pipeline((
                ("scaler", StandardScaler()),
                ("svr", LinearSVR(epsilon=1.5))
))
svm_reg.fit(X, y)
```

```python
def find_support_vectors(svm_reg, X, y):
    y_pred = svm_reg.predict(X)
    off_margin = (np.abs(y - y_pred) >= svm_reg.named_steps['svr'].epsilon)
    return np.argwhere(off_margin)

svm_reg.support = find_support_vectors(svm_reg, X, y)
```

```python
def plot_svm_regression(svm_reg, X, y, axes):
    x1s = np.linspace(axes[0], axes[1], 100).reshape(100, 1)
    y_pred = svm_reg.predict(x1s)
    plt.plot(x1s, y_pred, "k-", linewidth=2, label=r"$\hat{y}$")
    plt.plot(x1s, y_pred + svm_reg.named_steps['svr'].epsilon, "k--")
    plt.plot(x1s, y_pred - svm_reg.named_steps['svr'].epsilon, "k--")
    plt.plot(X, y, "bo")
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.ylabel(r"$y$", fontsize=18, rotation=0)
    plt.legend(loc="upper left", fontsize=18)
    plt.axis(axes)

plt.figure(figsize=(9, 4))
plot_svm_regression(svm_reg, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg.named_steps['svr'].epsilon), fontsize=18)
plt.scatter(X[svm_reg.support], y[svm_reg.support], s=180, facecolors='#FFAAAA')

plt.show()
```

```python
from sklearn.svm import SVR


np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()

svm_reg_poly = Pipeline((
                ("scaler", StandardScaler()),
                ("svr", SVR(kernel='poly', degree=2, C=100, epsilon=0.1))
))
svm_reg_poly.fit(X, y)
```

```python
plt.figure(figsize=(9, 4))
plot_svm_regression(svm_reg_poly, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_reg_poly.named_steps['svr'].degree, svm_reg_poly.named_steps['svr'].C, svm_reg_poly.named_steps['svr'].epsilon), fontsize=18)
plt.show()
```

### Under the Hood

#### Training Objective

Decision Function: f(x) = w.x + b, where:
   x is the input feature vector.
   w is the weight vector.
   b is the bias term.
   

Norm of the weight vector ||w|| represents the length (magnitude) of the weight vector. Slope of the decision function is equal to ||w||.

Margin is the distance between boundary (hyperplane) and the closest data points (support vectors). A larger margin generally indicates a better generalized model.
The margin is inversely proportional to the norm of the weight vector||w||. Minimizing ||w|| maximizes the margin.


Decision Boundary: The line where the decision function f(x) = 0.

Margins: The lines where the decision function equals +-1.


If we divide the weight vector by 2, the norm of the weight vector also gets halved. As margin is doubled, the points where the decision function equals +-1 are now at a distance 2/||w||, improving the model.


##### For Hard Margin
We also need to avoid any margin violation (hard margin), then we need the decision function to be greater than 1 for all position instances, and lower than -1 for negative training instances. 

If we define t(i) = -1 for negative instances (y(i) = 0) and t(i) = 1 for positive instances (y(i) = 1), then we can express this constraint as t(i)(wT.x(i) + b) >= 1 for all instances. 

As constrained Optimazition Problem:

Min(w, b) (1/2)wT.w

subject to: t(i)(wT.x(i) + b) >= 1   for i=1, 2, 3, ..., m


##### For Soft Margin
We need to introduce a slack varible  ζ(i) ≥ 0 (zeta) for each instance.  ζ(i) measures how much the ith instance is allowed to violate the margin. We need to make the slack varibales as small as possible to reduce the margin violations, and making (1/2)wT.w as small as possible to increase the margin.

C hyperparameter helps in this case, allowing us to define the trade-off between these two objectives.

Constrained Optimization Problem:


Min(w, b, ζ) (1/2)wT.w + C.∑(i=0)(m) ζ(i)

subject to: t(i)(wT.x(i) + b) >= 1 -ζ(i) and ζ(i) >=0  for i=1, 2, 3, ..., m


Taking small C, emphasizes margin maximization over the margin violations, leading to a wider margin but may allow more margin violations.

Taking large C, emphasizes margin violation minimization over margin maximization, leading to a narrower margin but fewer margin violations.

#### Quadratic Programming

Minimize(p) (1/2)pT.H.p + fT.p

Subject to A.p<=b

where, p = np-dimensional vector (np = number of parameters),
       H is an np x np matrix,
       f is an np-dimensional vector,
       A is an nc x np matrix (nc = number of constraints),
       b is an nc-dimensional vector
       

### Kernelized SVM

If we want to apply a 2nd-degree polynomial transformation to a two-dimensional training set, then train a linear SVM classifier on the transformed training set. 

2nd-degree polynomial mapping function:

f(x) = f((x1 x2)) = (x1^2  root2*x1*x2 x2^2)

The dot product of transformed vectors is equal to the square of the dot product of the original vectors: f(a)T . f(b) = (aT.b)^2.

In ML, a kernel is a function capable of computing the dot product f(a)T. f(b) based only on the original vectors a and b, without having to compute the transformation f. 

Common Kernels: Linear: K(a, b) = aT.b

Polynomial: K(a, b) = ( γaT ·b+r)^d

Guassian RBF: K(a, b) = exp(-γ ||a-b||^2)

Sigmoid: K(a, b) = tanh(γaT.b + r)

#### Hinge Loss

The function max(0, 1-t) is called the hinge loss function. It is equal to 0 when t>=1. Its derivative is equal to -1 if t<1 and 0 if t>1. It is not differentiable at t=1.

#### 7.  Train a LinearSVC on a linearly separable dataset. Then train an SVC and a SGDClassifier on the same dataset. See if you can get them to produce roughly the same model.

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler

c = 5
alpha = 1 / (c * len(X))
lin_clf = LinearSVC(loss="hinge", C=c, random_state=42)
svm_clf = SVC(kernel="linear", C=c)
sgd_clf = SGDClassifier(loss="hinge", learning_rate="constant", eta0=0.001, alpha=alpha,
                       max_iter=100000, tol=np.infty, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lin_clf.fit(X_scaled, y)
svm_clf.fit(X_scaled, y)
sgd_clf.fit(X_scaled, y)

print("LinearSVC: ", lin_clf.intercept_, lin_clf.coef_)
print("SVC: ", svm_clf.intercept_, svm_clf.coef_)
print("SGDClassifier: ", sgd_clf.intercept_, sgd_clf.coef_)
```

```python
w1 = -lin_clf.coef_[0, 0] / lin_clf.coef_[0, 1]
b1 = lin_clf.intercept_[0] / lin_clf.coef_[0, 1]

w2 = -svm_clf.coef_[0, 0] / svm_clf.coef_[0, 1]
b2 = svm_clf.intercept_[0] / svm_clf.coef_[0, 1]

w3 = -sgd_clf.coef_[0, 0] / sgd_clf.coef_[0, 1]
b3 = sgd_clf.intercept_[0] / sgd_clf.coef_[0, 1]

line1 = scaler.inverse_transform([[-10, -10 * w1 + b1], [10, 10 * w1 + b1]])
line2 = scaler.inverse_transform([[-10, -10 * w2 + b2], [10, 10 * w2 + b2]])
line3 = scaler.inverse_transform([[-10, -10 * w3 + b3], [10, 10 * w3 + b3]])


plt.figure(figsize=(11, 4))
plt.plot(line1[:, 0], line1[:, 1], "k:", label="LinearSVC")
plt.plot(line2[:, 0], line2[:, 1], "b--", linewidth=2, label="SVC")
plt.plot(line3[:, 0], line3[:, 1], "r-", label="SGDClassifier")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs") # label="Iris-Versicolor"
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo") # label="Iris-Setosa"
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper center", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.show()
```
