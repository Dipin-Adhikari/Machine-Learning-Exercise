```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif', 'color':'blue', 'size':20}
plt.title("Sports Watch Data", fontdict=font1, loc='left')
plt.xlabel("AveragePulse", fontdict=font1)
plt.ylabel("Calorie Burnage", fontdict=font1)

plt.plot(x, y)
plt.show()
```

```python
plt.xlabel('AveragePulse')
plt.ylabel('CalorieMeter')
plt.title('Sports Watch Data')
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.xlabel('AveragePulse')
plt.ylabel('CalorieMeter')
plt.title('Sports Watch Data')
plt.plot(x, y)
plt.grid(axis='x')
plt.show()
```

```python
plt.xlabel('AveragePulse')
plt.ylabel('CalorieMeter')
plt.title('Sports Watch Data')
plt.plot(x, y)
plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 2, 10])
plt.subplot(1, 2, 1)
plt.plot(x, y)

x = np.array([3, 2, 1, 6])
plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()

# subplot() takes three arguments, first and second is of row and column respectively and the third one is of the index of the plot.
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 2, 10])
plt.subplot(2, 1, 1)
plt.plot(x, y)

x = np.array([3, 2, 1, 6])
plt.subplot(2, 1, 2)
plt.plot(x, y)

plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 2, 10])
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Home')

x = np.array([3, 2, 1, 6])
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('Office')

plt.suptitle('Total')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 100, size=(10))
y = np.random.randint(1, 100, size=(10))
plt.scatter(x, y, color='hotpink')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 100, size=(100))
y = np.random.randint(1, 100, size=(100))
colors = np.random.randint(1, 100, size=(100))
sizes = np.random.randint(1, 100, size=(100))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.show()
```

```python
y = np.array([35, 25, 21, 15])
lbls = ['A', 'B', 'C', 'D']

plt.pie(y, labels=lbls)
plt.show()
```

```python
y = np.array([35, 25, 21, 15])
lbls = ['A', 'B', 'C', 'D']
explodes = [0.2, 0, 0, 0]
plt.pie(y, labels=lbls, startangle=90, explode=explodes, shadow=True)
plt.legend(title='Fruits')
plt.show()
```
